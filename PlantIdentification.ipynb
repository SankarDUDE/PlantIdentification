```python
#creating dictionary of flower species
flower_images_dict={
    "Apple___Black_rot" :list(directory.glob('Apple___Black_rot/*.jpg')),
    "Apple___healthy" :list(directory.glob('Apple___healthy/*.jpg')),
    "Blueberry___healthy" :list(directory.glob('Blueberry___healthy/*.jpg')),
    "Cherry_(including_sour)___healthy" :list(directory.glob('Cherry_(including_sour)___healthy/*.jpg')),
    "Cherry_(including_sour)___Powdery_mildew" :list(directory.glob('Cherry_(including_sour)___Powdery_mildew/*.jpg')),
    "Corn_(maize)___Common_rust_" :list(directory.glob('Corn_(maize)___Common_rust_/*.jpg')),
    "Corn_(maize)___healthy" :list(directory.glob('Corn_(maize)___healthy/*.jpg')),
    "Grape___Black_rot" :list(directory.glob('Grape___Black_rot/*.jpg')),
    "Grape___healthy" :list(directory.glob('Grape___healthy/*.jpg')),
    "Peach___healthy" :list(directory.glob('Peach___healthy/*.jpg')),
    "Pepper,_bell___Bacterial_spot" :list(directory.glob('Pepper,_bell___Bacterial_spot/*.jpg')),
    "Pepper,_bell___healthy" :list(directory.glob('Pepper,_bell___healthy/*.jpg')),
    "Potato___Early_blight" :list(directory.glob('Potato___Early_blight/*.jpg')),
    "Potato___healthy" :list(directory.glob('Potato___healthy/*.jpg')),
    "Strawberry___healthy" :list(directory.glob('Strawberry___healthy/*.jpg')),
    "Tomato___Bacterial_spot" :list(directory.glob('Tomato___Bacterial_spot/*.jpg')),
    "Tomato___healthy" :list(directory.glob('Tomato___healthy/*.jpg')),
}
flower_images_dict
```

```python
#Different species list
keys=["Apple___Black_rot","Apple___healthy","Blueberry___healthy","Cherry_(including_sour)___healthy","Cherry_(including_sour)___Powdery_mildew","Corn_(maize)___Common_rust_","Corn_(maize)___healthy","Grape___Black_rot","Grape___healthy","Peach___healthy","Pepper,_bell___Bacterial_spot","Pepper,_bell___healthy","Potato___Early_blight","Potato___healthy","Strawberry___healthy","Tomato___Bacterial_spot","Tomato___healthy"]
```

```python
#resizing and creating labels using computer vision
resized,labels=[],[]
for flower_name,images in flower_images_dict.items():
    for image in images:
        print("completed")
        img=cv2.imread(str(image))
        resized_image=cv2.resize(img,(224,224))
        resized.append(resized_image)
        labels.append(keys.index(flower_name))
flower_images_dict
```

```python
#function to print images
def print_image(i,j):
    plt.imshow(i)
    plt.title(keys[j])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(resized,labels,test_size=0.25,random_state=0)
```

```python
print_image(x_train[15],y_train[15])
```

```python
print_image(x_train[150],y_train[150])
```

```python
x_train_scaled=np.array(x_train)/255
x_test_scaled=np.array(x_test)/255
```

```python
x_train_scaled.shape
```

```python
y_train=np.array(y_train)
y_train.shape
```

SIMPLE CNN MODEL

```python
model=models.Sequential([
    layers.Conv2D(16,3,padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128,activation="relu"),
    layers.Dense(17,activation="softmax")])
model.compile(
optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train_scaled,np.array(y_train),epochs=10)
```

```python
model.evaluate(x_test_scaled,np.array(y_test))
```

```python
model1=models.Sequential([
    layers.experimental.preprocessing.RandomZoom(0.3),
    layers.Conv2D(16,3,padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128,activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(64,activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(17,activation="softmax")
    #layers.Dense(2,kernel_regularizer=l2(0.01),activation="softmax")
])
model1.compile(
optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

CNN MODEL THAT USES FILTERS IN ALL THE **LAYERS**

```python
myCNN=tf.keras.models.Sequential([
    layers.BatchNormalization(),
    layers.Conv2D(32,3,activation="relu"),
    layers.MaxPooling2D(),
    layers.Conv2D(64,3,activation="relu"),
    layers.MaxPooling2D(),
    layers.Conv2D(128,3,activation="relu"),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(256,activation="relu"),
    layers.Dense(17,activation="softmax")    
])
myCNN.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
myCNN.fit(x_train_scaled,np.array(y_train),epochs=5)
```
